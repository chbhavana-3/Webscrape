```
def get_seller_info(soup):
  s_inf=soup.find_all('span',attrs={'class':'seller-type'})
  s_info=[s.text for s in s_inf]
  return s_info
get_seller_info(soup)
```

```
def get_Locality(soup):
  loc= soup.find_all('span',attrs={'itemprop':"addressLocality"})
  locality=[i.text for i in loc]
  return locality
get_Locality(soup)
```

```
def get_area(soup):
  ar= soup.find_all('td',attrs={'class':'size'})
  area=[i.text for i in ar]
  return area
get_area(soup)
```

```
def get_price(soup):
  b_info=soup.find_all('td',attrs={'class':"price"})
  c_info=[b.text for b in b_info]
  return c_info
get_price(soup)
```

```
def get_no_bathrooms(soup):
   bh=soup.find_all('ul',attrs={'class':'listing-details'})
   b_rooms=[]
   for j in bh:
     try:
       b_rooms.append(j.li.text.split(' ')[0])
     except AttributeError:
       b_rooms.append('None')
   return b_rooms
get_no_bathrooms(soup)
```

```
def makaanscrape(url):
  request=requests.get(url)
  soup=BeautifulSoup(request.content,'html')
  area=get_area(soup)
  locality=get_Locality(soup)
  bath=get_no_bathrooms(soup)
  price=get_price(soup)

  return area,locality,bath,price
```

```

area,locality,bath,price=makaanscrape('https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city')
print(area)
print(locality)
print(bath)
print(price)
```
